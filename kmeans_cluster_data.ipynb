```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans,DBSCAN
from sklearn import metrics
```

```python
data=pd.read_csv("cluster_data.csv")
data.head()
```

```python
wcss=[]
kl=range(2,11)
for i in kl:
  k=KMeans(n_clusters=i)
  k=k.fit(data)
  wcss.append(k.inertia_)
```

```python
plt.plot(kl,wcss,"*-")
plt.show()
```

```python
from sklearn.decomposition import PCA
p=PCA(n_components=2)
p_d=p.fit_transform(data)
k=KMeans(n_clusters=4)
x=k.fit_predict(data)
```

```python
plt.scatter(p_d[:,0],p_d[:,1],c=x)
plt.xlabel("pca1")
plt.ylabel("pca2")
plt.show()
```

```python
plt.scatter(k.cluster_centers_[:, 0], k.cluster_centers_[:, 1], c='red', marker='X', s=200, label='Centroids')
```
