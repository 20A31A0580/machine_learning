```python

import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
cluster_DATA = pd.read_csv('cluster_data.csv')
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(cluster_DATA)
```

```python
dbscan = DBSCAN(eps=3.5, min_samples=4)
cluster_labels = dbscan.fit_predict(scaled_features)
```

```python
plt.scatter(cluster_DATA.iloc[:, 0], cluster_DATA.iloc[:, 1], c=cluster_labels, cmap='viridis', edgecolors='k', s=50)
plt.title('DBSCAN Clustering')
plt.show
```

```python
from sklearn.decomposition import PCA
p=PCA(n_components=2)
p_d=p.fit_transform(cluster_DATA)
plt.scatter(p_d[:,0],p_d[:,1],c=cluster_labels)
plt.xlabel("pca1")
plt.ylabel("pca2")
plt.show()
```
